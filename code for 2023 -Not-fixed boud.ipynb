```python
from mip import Model, xsum, maximize, BINARY
import numpy as np


#Randomly typed
preferences_matrix = preferences_matrix2023


project_types = project_types2023
lower_bounds = lower_bounds_2023
upper_bounds = upper_bounds_2023


model = Model()

# Decision variables
x = [[model.add_var(var_type=BINARY) for j in range(N_s)] for i in range(N_p)]


# Objective Function

model.objective = maximize(
     xsum(x[i][j] * preferences_matrix[j][i] for i in range(N_p) for j in range(N_s)))


# Constraint: enforce lower bound and upper bound for each project 
for i in range(N_p):
    model += xsum(x[i][j] for j in range(N_s)) >= lower_bounds[i]
    model += xsum(x[i][j] for j in range(N_s)) <= upper_bounds[i]
    
model.optimize()

# Initialize the allocation matrix with zeros
allocation_matrix = np.zeros((N_p, N_s))

# Track whether a student has been allocated
student_allocated = [False] * N_s

# Extract and print the solution
if model.num_solutions:
    for i in range(N_p):
        for j in range(N_s):
            if x[i][j].x >= 0.99 and not student_allocated[j] and preferences_matrix[j, i] != -200:
                allocation_matrix[i, j] = 1
                student_allocated[j] = True  # Mark this student as allocated
                print(f"Student {j} is allocated to Project {i+1}")

# Allocation Matrix
print("Allocation Matrix (Projects x Students):")
print(allocation_matrix)

# Initialize a variable to hold the total score
total_score = 0

# Iterate over the allocation matrix to calculate the score
for i in range(N_p):
    for j in range(N_s):
        if allocation_matrix[i, j] == 1:  # If this student is allocated to this project
            # Add the preference score for this allocation to the total score
            total_score += preferences_matrix[j, i]    

# Print the total score
print(f"Total Preference Score: {total_score}")
print(model.objective_value)

# Check if every student is allocated to at least one project
for j in range(N_s):
    if not student_allocated[j]:
        print(f"Student {j} is not allocated to any project.")

        
```

```python
# Initialize a set to store allocated students
allocated_students = set()

# Flag to indicate if any repeating occurrence is found
repeating_occurrence_found = False

# Iterate through each project-student allocation
for i in range(N_p):
    for j in range(N_s):
        if allocation_matrix[i, j] == 1:
            # Check if the student has already been allocated
            if j in allocated_students:
                print(f"Student {j} is allocated to multiple projects.")
                repeating_occurrence_found = True
            else:
                # Add the student to the set of allocated students
                allocated_students.add(j)

# If no repeating occurrences are found, print a message
if not repeating_occurrence_found:
    print("No repeating occurrences of students being allocated to multiple projects.")
```
